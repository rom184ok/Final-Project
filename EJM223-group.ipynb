```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
#goes through dataset
data = pd.read_csv("C:\\Users\\grape\\Downloads\\fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv")
count = data.groupby('neighborhood').size()
#turns into column and renames
count = count.reset_index(name = 'count')
#reads through the map file
neighborhood = gpd.read_file("C:\\Users\\grape\\Downloads\\neighborhoods\\Neighborhoods_.shp")
#matches the columns for hood to the ones from neighborhood to connect the shapefile
neighborhood = neighborhood.merge(count, left_on = 'hood', right_on = 'neighborhood', how = 'left')
#makes 8 by 8 fig and ax
fig, ax = plt.subplots(1, 1, figsize = (8,8))
#gets the color through count, adds a legend to the side and labels it Total Number of Parks and have it in shades of blue with black edges
neighborhood.plot(column = 'count',ax=ax, legend = True, legend_kwds= {"label": "Total Number of Parks"},cmap = 'Blues', edgecolor = 'black')
#axis wouldnt make sense 
ax.set_axis_off()
#prints
plt.show()
```
